```python
data = pd.read_csv("evolution_time_results.csv")
data = data[data['Project id'] == 129181598]
data = data.reset_index(drop=True)
data.head()
```

```python
data = data[["Project date","Block Count"]]
data.shape
```

```python
data2 = data
data2["test"] = ""
data2

block_usage_cleaned = []
for i in data2.index:
    data2["Block Count"][i] = data2["Block Count"][i].strip("[]")
    data2["Block Count"][i] = data2["Block Count"][i].split(",")
    array = [item.replace(" '  ","").replace("'  ","").replace(" ' ","").replace("'","") for item in data2["Block Count"][i]]
    dic = {}
    for item in array:
        key = re.sub("\d ","",item)
        key = re.sub("\d","",key)
        key = re.sub("\d","",key)
        key = re.sub("\d","",key)
        key = re.sub("\d","",key)
        key = re.sub("\d","",key)
# not an elegent way but effective lol: adding multiple key = re.sub("\d","",key)
        value = item.split(" ")[0]
        dic[key] = value     
    block_usage_cleaned.append(dic)
    data2["test"][i] = dic
    i = i + 1
    
len(block_usage_cleaned)
```
